```python
import pandas as pd
import matplotlib.pyplot as plt

dados = pd.read_csv('data.csv')

idades = dados['Age']
salario_devs = dados['All_Devs']
py_salario = dados['Python']
js_salario = dados['JavaScript']

plt.plot(idades, salario_devs, color="black", linestyle='--', label='Todos os Devs')
plt.plot(idades, py_salario, label='Python Dev')

plt.fill_between(idades, py_salario, alpha=0.25)#alpha serve para almentar ou diminuir a transparencia
plt.legend()
plt.title('Média Salarial(USO) por Idade')
plt.xlabel('Idade')
plt.ylabel('Média salarial (USO)')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('seaborn')
dados = pd.read_csv('2019-05-31-data.csv')

views = dados['view_count']
likes = dados['likes']
ratio = dados['ratio']

plt.scatter(views, likes, edgecolors='black', c=ratio, cmap='summer', linewidths=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')
plt.xscale('log')
plt.yscale('log')
plt.title("Treending Youtube Videos")
plt.xlabel("View Count")
plt.ylabel("Total likes")
plt.tight_layout()
```

```python
import pandas as pd
from matplotlib import pyplot as plt

plt.style.use('seaborn')

data = pd.read_csv('data.csv')
ages = data['Age']
dev_salaries = data['All_Devs']
py_salaries = data['Python']
js_salaries = data['JavaScript']

fig, (ax1, ax2, ax3) = plt.subplots(nrows=3, ncols=1, sharex=True)

ax1.plot(ages, dev_salaries, color='#444444',
         linestyle='--', label='All Devs')

ax2.bar(ages, py_salaries, label='Python', color='pink')

ax3.plot(ages, js_salaries, label='JavaScript', color='black')


ax1.legend()

ax1.set_title('Median Salary (USD) by Age')
#ax1.set_xlabel('Ages')
ax1.set_ylabel('Median Salary (USD)')

ax2.legend()

#ax2.set_title('Median Salary (USD) by Age')
#ax2.set_xlabel('Ages')
ax2.set_ylabel('Median Salary (USD)')

ax3.legend()

#ax3.set_title('Median Salary (USD) by Age')
ax3.set_xlabel('Ages')
ax3.set_ylabel('Median Salary (USD)')


plt.tight_layout()

plt.show()
```
